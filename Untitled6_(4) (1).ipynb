```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dialogues = []
with open('movie_lines.txt', 'r', encoding='utf-8', errors='ignore') as f:
    for line in f:
        # Split by the separator " +++$+++ "
        parts = line.split(" +++$+++ ")
        if len(parts) == 5:
            # The text is the last part (index 4)
            text = parts[4].strip()
            dialogues.append(text)

print(f"Loaded {len(dialogues)} lines of dialogue.")
```

```python
glove_embedding_dim = 100

embeddings_index = {}
with open('glove.6B.100d.txt', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs

print(f'Found {len(embeddings_index)} word vectors.')
```

```python
data_subset = dialogues[:120000]
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocab_size= 15000 # perfect to reduce noise and rare words
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
tok  = Tokenizer(num_words = vocab_size,oov_token = "<OOV>")
tok.fit_on_texts(data_subset)
seqs = tok.texts_to_sequences(data_subset)
```

```python
X = []
y=[]
for line in seqs:
    for i in range(1, len(line)):
        n_gram_sequence = line[:i]
        out_seq= line[i]
        X.append(n_gram_sequence)
        y.append(out_seq)
```

```python
max_seq_len = 50 #avg len to comply with hardware capibilities
X_padded = pad_sequences(X, maxlen=max_seq_len, padding='pre')
X_final = np.array(X_padded)
y_final = np.array(y)
```

```python
embedding_matrix = np.zeros((vocab_size, glove_embedding_dim))
for word, i in tok.word_index.items():
    if i < vocab_size:
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
        # Words not found in embedding index will be all-zeros.

print(f'Embedding matrix shape: {embedding_matrix.shape}')
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SimpleRNN, Dropout, Bidirectional,Input
```

```python
embed_dim = glove_embedding_dim # Update embed_dim to match GloVe dimensions
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

```python
lstm = Sequential([
    Input(shape=(max_seq_len,)),
    Embedding(input_dim=vocab_size, output_dim=embed_dim, weights=[embedding_matrix], trainable=False),
    LSTM(256, return_sequences=True),
    Dropout(0.3),
    LSTM(128),
    Dropout(0.2),
    Dense(vocab_size, activation='softmax')
])

lstm.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print(lstm.summary())
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint(
    'model_checkpoint.keras',
    monitor='val_loss',
    save_best_only=True,
    mode='min'
)
```

```python
history= lstm.fit(shuffle=True, x=X_final,y=y_final,batch_size=128,epochs=100,verbose=1,validation_split=0.1,callbacks=[early_stop,checkpoint])
```

```python
lstm.save("movie_lstm.h5")
```

```python
import numpy as np

def get_suggestions(seed_text, model, tokenizer, max_sequence_len, top_n=3):
    """Returns the top N most likely next words."""
    # 1. Preprocessing the seed text
    seed_text_cleaned = clean_text(seed_text.lower())
    token_list = tokenizer.texts_to_sequences([seed_text_cleaned])[0]
    token_list = pad_sequences([token_list], maxlen=max_sequence_len, padding='pre')

    # 2. Get prediction probabilities
    predicted_probs = model.predict(token_list, verbose=0)[0]

    # 3. Ignore index 0 (padding) and index 1 (OOV)
    predicted_probs[0] = 0
    predicted_probs[1] = 0

    # 4. Get indices of the top N probabilities
    top_indices = np.argsort(predicted_probs)[-top_n:][::-1]

    # 5. Map indices back to words
    suggestions = []
    for index in top_indices:
        for word, word_index in tokenizer.word_index.items():
            if word_index == index:
                suggestions.append(word)
                break
    return suggestions

def generate_text_no_oov(seed_text, next_words, model, max_sequence_len, tokenizer, temperature=1.0):
    """Generates text while explicitly skipping the OOV token."""
    output_text = seed_text

    for _ in range(next_words):
        cleaned_seed = clean_text(output_text.lower())
        token_list = tokenizer.texts_to_sequences([cleaned_seed])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len, padding='pre')

        predicted_probs = model.predict(token_list, verbose=0)[0]

        # --- FIX OOV: Set probability of OOV (index 1) and Padding (index 0) to zero ---
        predicted_probs[0] = 0
        predicted_probs[1] = 0

        # Apply temperature and re-normalize
        predicted_probs = np.asarray(predicted_probs).astype('float64')
        predicted_probs = np.log(predicted_probs + 1e-10) / temperature
        exp_preds = np.exp(predicted_probs)
        predicted_probs = exp_preds / np.sum(exp_preds)

        # Sample the next word
        predicted_word_index = np.random.choice(len(predicted_probs), p=predicted_probs)

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                output_word = word
                break

        if output_word == "" or output_word == "<OOV>":
            continue # Skip if it still hits OOV

        output_text += " " + output_word

    return output_text

# --- TESTING ---

# 1. Get Autocomplete Suggestions
seed = "I want to have this"
top_3 = get_suggestions(seed, lstm, tok, max_seq_len, top_n=3)
print(f"Suggestions for '{seed}': {top_3}")

# 2. Generate a sentence without OOV
generated = generate_text_no_oov("the time is not in our", 5, lstm, max_seq_len, tok, temperature=0.7)
print(f"Generated sentence: {generated}")
```

```python
import numpy as np

def beam_search_predictions(seed_text, model, tokenizer, max_sequence_len, beam_width=3, next_words=5):
    # Start with a list of (sequence, probability)
    # The probability is stored in log space to prevent underflow (multiplying small decimals)
    start_seq = tokenizer.texts_to_sequences([seed_text])[0]
    candidates = [(start_seq, 0.0)]

    for _ in range(next_words):
        all_candidates = []

        # Expand each current candidate path
        for seq, score in candidates:
            padded_seq = pad_sequences([seq], maxlen=max_sequence_len-1, padding='pre')
            probs = model.predict(padded_seq, verbose=0)[0]

            # Mask Padding and OOV (indices 0 and 1)
            probs[0] = 1e-10
            probs[1] = 1e-10

            # Get Top K indices
            best_k = np.argsort(probs)[-beam_width:]

            for idx in best_k:
                new_seq = seq + [idx]
                # Log probabilities turn multiplication into addition
                new_score = score + np.log(probs[idx] + 1e-10)
                all_candidates.append((new_seq, new_score))

        # Sort all possible paths by score and keep top 'beam_width'
        ordered = sorted(all_candidates, key=lambda x: x[1], reverse=True)
        candidates = ordered[:beam_width]

    # Convert the best path back to words
    best_path = candidates[0][0]
    output_words = [tokenizer.index_word.get(idx, '<OOV>') for idx in best_path]
    return " ".join(output_words)

# --- TEST IT ---
result = beam_search_predictions("the time is not in our", lstm, tok, max_seq_len, beam_width=3, next_words=5)
print(f"Beam Search Result: {result}")
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9' ]", '', text) # Remove punctuation except for apostrophes
    return text
```
